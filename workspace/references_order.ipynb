```python
# Dataset https://www.aminer.org/citation
import pandas as pd
import json

with open('data/dblp-ref-3.json', 'r') as f:
    records = f.read().split('\n')[:-1]
    data = '['+','.join(records)+']'
    df = pd.read_json(data)
df.head(10)
```

```python
df.iloc[0].references
```

```python
def get_citation(references):
    try:
        return list(df[df.id.isin(references)].id.values)
    except:
        return []
df['citations'] = df.references.apply(get_citation)
```
