```python
import pandas as pd

def to_int(x):
    try:
        return int(x)
    except Exception:
        print(x)
        return 0

df = pd.read_csv('text_clustering.csv')\
    .dropna(subset=['Author Keywords'])

def norm(kw):
    kw = kw.strip()
    if len(kw) == 0:
        return None
    return kw

df['Author Keywords'] = df['Author Keywords'].apply(lambda x: [norm(kw) for kw in str(x).lower().split(';') if norm(kw) is not None])
df_cnt = df[['Year', 'Author Keywords']]\
    .rename(columns={'Author Keywords': 'Count'})\
    .groupby('Year')\
    .count()\
    .reset_index()
df_kw_by_year = df[['Year', 'Author Keywords', 'Titles']]\
    .rename(columns={'Titles': 'Count'})\
    .explode('Author Keywords')\
    .groupby(['Year', 'Author Keywords'])\
    .count()\
    .reset_index()
df_kw_by_year = df_kw_by_year.sort_values(['Year','Count'], ascending=False).groupby('Year').head(10)
df_kw_by_year['rank'] = 10 - df_kw_by_year.groupby('Year').cumcount()
# df_kw_by_year['Count'] = df_kw_by_year.apply(lambda x: to_int(x['Count']) / to_int(df_cnt[df_cnt['Year']==x['Year']]['Count'].values[0]), axis=1)
```
